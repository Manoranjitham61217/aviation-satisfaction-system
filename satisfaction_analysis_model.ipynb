```python
cat_features = df.select_dtypes(include=["object"]).columns
num_features = df.select_dtypes(include=["int64", "float64"]).columns
```

```python
num_features
```

```python
cat_features
```

```python
from sklearn.preprocessing import LabelEncoder
encoder={}
for col in df[cat_features]:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
    encoder[col]=le
```

```python
df.head()
```

```python
x = df.drop('satisfaction', axis=1)
y = df['satisfaction']
```

```python
from sklearn.feature_selection import SelectKBest,chi2,f_regression
x_cat = df[cat_features]

chi_selector = SelectKBest(score_func=chi2, k='all')
chi_selector.fit(x_cat, y)

chi_scores = pd.DataFrame({
    'Feature': cat_features,
    'Chi2 Score': chi_selector.scores_
}).sort_values(by='Chi2 Score', ascending=False)

print(chi_scores)
```

```python
chi2_selected = x_cat.columns[chi_selector.get_support()].tolist()
```

```python
x_num = df[num_features]

f_selector = SelectKBest(score_func=f_regression, k=10)
f_selector.fit(x_num, y)

f_scores = pd.DataFrame({
    'Feature': num_features,
    'Fisher\'s Score': f_selector.scores_
}).sort_values(by='Fisher\'s Score', ascending=False)

print(f_scores)
```

```python
freg_selected = x_num.columns[f_selector.get_support()].tolist()
```

```python
selected_features = list(set(chi2_selected + freg_selected))
selected_features
```

```python
x_full = pd.concat([x_num, x_cat], axis=1)
```

```python
x_full
```

```python
x_selected = x_full[selected_features]
```

```python
x_selected
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_selected,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import time

start=time.time()
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
end=time.time()

y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Calssification Report: \n",classification_report(y_test,y_pred))
print("Time Taken: ",end-start)
model.n_iter_
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10,5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Purples")
plt.title("Confusion Matrix - LogisticRegression(without Scaling)")
plt.show()
```

```python
from sklearn.model_selection import  StratifiedKFold, RandomizedSearchCV
```

```python
start=time.time()
```

```python
xgb = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    random_state=42
)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300, 500, 800],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7, 10],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.5, 0.7, 1.0],
    'gamma': [0, 0.1, 0.3, 1],
    'min_child_weight': [1, 3, 5]
}
```

```python
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

```python
random_search_xgb = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    n_iter=10,
    scoring='accuracy',
    cv=kfold,
    verbose=1,
    n_jobs=-1,
    random_state=42
)
```

```python
model_xgb = random_search_xgb.fit(x_selected, y)
end=time.time()
```

```python
print("Best Hyperparameters:\n", model_xgb.best_params_)
print("\nBest Cross Validation Accuracy:", model_xgb.best_score_)
```

```python
y_pred = model_xgb.predict(x_selected)

print("\nAccuracy:", accuracy_score(y, y_pred))
print("\nClassification Report:\n", classification_report(y, y_pred))
print("\nTime Taken: ",(end-start))
```

```python
cm = confusion_matrix(y, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Purples")
plt.title("Confusion Matrix - XGBoost(With Hyperparameter Tuning)")
plt.show()
```
